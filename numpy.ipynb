1. Remove Duplicates from 1D Array

````python
import numpy as np

arr = np.array([1, 2, 2, 3, 4, 4, 5])

unique_arr = np.unique(arr)

print(unique_arr)
````

2. Remove Duplicate Rows from 2D Array

````python
import numpy as np

arr = np.array([[1, 2],
                [3, 4],
                [1, 2],
                [5, 6]])

unique_rows = np.unique(arr, axis=0)

print(unique_rows)
````

3. Remove Duplicate Columns

````python
unique_cols = np.unique(arr, axis=1)
print(unique_cols)
````

4. Get Unique Values + Counts

````python
values, counts = np.unique(arr, return_counts=True)

print("Values:", values)
print("Counts:", counts)
````

5. Get Unique Values + Index

````python
values, index = np.unique(arr, return_index=True)

print("Unique values:", values)
print("First occurrence index:", index)
````
