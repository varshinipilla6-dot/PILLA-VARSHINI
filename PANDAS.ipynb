Step 1: Load the Dataset

```python
import pandas as pd
df = pd.read_csv('/swiggy.csv')
```

Step 2: Understand the Data

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

Step 3: Remove Duplicate Values

```python
df = df.drop_duplicates()
```

Step 4: Handle Missing Values

```python
df.isnull().sum()
df = df.dropna()
```

Step 5: Clean Text Data

```python
df['restaurant_name'] = df['restaurant_name'].str.strip().str.lower()
```

```python
df['Restaurant'] = df['Restaurant'].str.strip().str.lower()
```

Step 6: Clean Cost Column

```python
df['cost'] = df['cost'].str.replace('₹', '')
```

```python
df['Price'] = df['Price'].astype(str).str.replace('₹', '')
```

```python
df['Price'] = df['Price'].astype(str).str.replace(' for two', '')
```

```python
df['Price'] = df['Price'].astype(float)
```

Step 7: Convert Data Types

```python
df['Avg ratings'] = pd.to_numeric(df['Avg ratings'], errors='coerce')
```

Step 8: Remove Invalid Data

```python
df = df[(df['Avg ratings'] > 0) & (df['Avg ratings'] <= 5)]
df = df[df['Price'] > 0]
```
