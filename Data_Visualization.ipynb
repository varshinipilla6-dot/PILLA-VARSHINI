1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2. Load Dataset

```python
df = pd.read_csv('/content/swiggy.csv')
```

3. Bar Chart (Top Cuisines)

```python
print(df.columns)
```

4. Pie Chart (City Distribution)

```python
city_count = df['City'].value_counts().head(5)

plt.pie(city_count, labels=city_count.index, autopct='%1.1f%%')
plt.title("City-wise Distribution")
plt.show()
```

5. Histogram (Ratings Distribution)

```python
plt.hist(df['Avg ratings'], bins=5)
plt.title("Average Rating Distribution")
plt.xlabel("Average Rating")
plt.ylabel("Frequency")
plt.show()
```

6. Scatter Plot (Cost vs Rating)

```python
plt.scatter(df['Price'], df['Avg ratings'])
plt.title("Price vs Average Rating")
plt.xlabel("Price")
plt.ylabel("Average Rating")
plt.show()
```

7. Line Chart (Average Cost by City

```python
avg_cost = df.groupby('City')['Price'].mean().sort_values().head(10)

avg_cost.plot(kind='line', marker='o')
plt.title("Average Price by City")
plt.xlabel("City")
plt.ylabel("Average Price")
plt.show()
```
